```python
import numpy as np
```

```python
theta_x = -10 * np.pi / 180
theta_y = -20 * np.pi / 180
theta_z = -15 * np.pi / 180

rotation_matrix_x = np.array([
    [1, 0, 0],
    [0, np.cos(theta_x), -np.sin(theta_x)],
    [0, np.sin(theta_x), np.cos(theta_x)]
])

rotation_matrix_y = np.array([
    [np.cos(theta_y), 0, np.sin(theta_y)],
    [0, 1, 0],
    [-np.sin(theta_y), 0, np.cos(theta_y)]
])

rotation_matrix_z = np.array([
    [np.cos(theta_z), -np.sin(theta_z), 0],
    [np.sin(theta_z), np.cos(theta_z), 0],
    [0, 0, 1]
])

R = rotation_matrix_x @ rotation_matrix_z @ rotation_matrix_y

yz_normal_vector = np.array([1, 0, 0])
rotated_yz_vector = R @ yz_normal_vector
print("Rotated YZ vector:", rotated_yz_vector)

xy_normal_vector = np.array([0, 0, 1])

cross_product = np.cross(rotated_yz_vector, xy_normal_vector)
print("Cross product:", cross_product)

cross_product = np.array([np.cos(theta_x)*np.sin(theta_z)*np.cos(theta_y) + np.sin(theta_x)*np.sin(theta_y),
                          -np.cos(theta_z)*np.cos(theta_y),
                          0])
print("calculated Cross product:", cross_product)

base_vector = np.array([0, -1, 0])

xy_theta = np.arccos(np.clip(np.dot(cross_product, base_vector) /
                          (np.linalg.norm(cross_product) * np.linalg.norm(base_vector)), -1.0, 1.0))

print("XY theta (degrees):", xy_theta * 180 / np.pi)
```
