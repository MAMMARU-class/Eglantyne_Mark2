```python
import numpy as np
```

```python
l_com_z = -52.2
l_com_y = 30

l_base_roll = 22.2
l_roll_pitch = 26.01
l_leg = 78.02

l_roll_com = l_com_z + l_base_roll

l_foot_z = 37.5
l_foot_x = 21
```

```python
com_order = [0,30,158,0,0]
angles = [0,0,0,0,0,0]
```

```python
com = [com_order[0], com_order[1], com_order[2]]
theta = com_order[3]
phi = com_order[4]
```

```python
# rotate
foot2leg_yaw = [0,0]
foot2leg_yaw[0] = com[0] + l_com_y * np.sin(theta)
foot2leg_yaw[1] = com[1] - l_com_y * np.cos(theta)
print("foot2leg_yaw:", foot2leg_yaw)

# leg_yaw
angles[0] = theta
print("leg_yaw:", angles[0])
```

```python
# back
alpha = np.arctan2(foot2leg_yaw[1], com[2] - (l_roll_com + l_foot_z))

# leg_roll, foot_roll
angles[1] = alpha
angles[5] = alpha
print("leg_roll:", angles[1])
print("foot_roll:", angles[5])
```

```python
# side
arctan_x = com[2] - (l_roll_com + l_foot_z + l_roll_pitch * np.cos(alpha))
arctan_y = foot2leg_yaw[0] + l_foot_x

beta = np.arctan2(arctan_y, arctan_x)
print("beta:", beta)
```

```python
# knee
l_leg_all = ( com[2] - (l_roll_com + l_foot_z + l_roll_pitch * np.cos(alpha)) ) / np.cos(alpha) / np.cos(beta)
gamma = np.arccos( (l_leg_all/2) / l_leg)

# leg_upper, leg_foot, leg_under
angles[2] = gamma + beta
angles[4] = gamma - beta
angles[3] = np.pi -2 * gamma
print("leg_upper:", angles[2])
print("foot_pitch:", angles[4])
print("leg_under:", angles[3])
```
